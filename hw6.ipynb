```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
ratio=1
state=[0,1] 
t = np.arange(0.0, 100, 0.01)
```

```python
def f1(state,t):
    theta,omega=state
    return omega,-ratio*np.sin(theta)
```

```python
sol1=odeint(f1,state,t)
plt.plot(t,sol1[:,0],label=r'$\theta (t)$')
plt.legend()
plt.ylabel(r'$ \theta$')
plt.xlabel('$t$')
plt.show()
```

```python
def f2(state,t):
    theta,omega=state
    return omega,-ratio*theta
```

```python
sol2=odeint(f2,[0,1],t)
plt.plot(t,sol2[:,0],label=r'$small oscillations$')
plt.legend()
plt.ylabel(r'$ \theta$')
plt.xlabel('$t$')
plt.show()
```
