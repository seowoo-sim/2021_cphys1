```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def leibniz(n):
    sum=0
    x=0
    while (2*x+1)<=n:
        sum+=(-1)**x/(2*x+1)
        x+=1
    return sum
```

```python
n=np.arange(1,100,2)
plt.plot(n,[leibniz(x) for x in n],label='Leibniz formula for $\pi$')
plt.plot(n,[np.pi/4 for x in range(50)],label='$\pi$/4')
plt.legend()
plt.xlabel('n')
plt.show()
```

```python
def logistic(x,r,f0):
    if x==0:
        return f0
    else:
        f=f0
        for i in range(x):
            f=r*f*(1-f)
        return f
```

```python
x1=np.arange(1,101,1)
```

```python
plt.plot(x1,[logistic(i,0.5,0.5) for i in x1])
plt.plot(x1,[logistic(i,0.95,0.5) for i in x1])
plt.plot(x1,[logistic(i,2,0.5) for i in x1])
plt.plot(x1,[logistic(i,3.2,0.5) for i in x1])
plt.plot(x1,[logistic(i,3.8,0.5) for i in x1])
plt.plot(x1,[logistic(i,3.8,0.501) for i in x1])
plt.show()
```

```python
def sinc(x):
    return np.sinc(x/np.pi)
```

```python
def sinc_tn(x,N):
    j=0
    sum=0
    while j<=N:
        if j%2==0:
            sum+=(x**(2*j))/np.math.factorial(2*j+1)
            j=j+1
        else:
            sum-=(x**(2*j))/np.math.factorial(2*j+1)
            j=j+1
    return sum
```

```python
n2=np.arange(1,10,0.0001)
plt.plot(n2,[sinc(i) for i in n2],label='sinc(x)')
plt.plot(n2,[sinc_tn(x,8) for x in n2],label='N=8')
plt.plot(n2,[sinc_tn(x,9) for x in n2],label='N=9')
plt.plot(n2,[sinc_tn(x,10) for x in n2],label='N=10')
plt.legend()
plt.show()
```
